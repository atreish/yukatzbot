```python
entities_list=[]
```

```python
for enitity in entities:
    entities_list.extend(enitity.split(','))

```

```python
entities_list=[x for x in entities_list if x!='\n']
```

```python
for i, enitity in enumerate(entities_list):
  if ('\\n' in enitity) or ('\n' in enitity) :
    del entities_list[i]

    entities_list.extend(enitity.split('\\n'))
```

Clean of signs in the begining of string

```python
entities_list
```

```python
regex_delete_number=r'\d+.'
regex_del_title=r'Dr.'
regex_del_quotion_mark_start=r"^('|\[')"
regex_del_quotion_mark_end=r"'$"
regex_del_quotion_period=r'\.$'
regex_del_quotion_dash=r'^(-|")'
```

```python
for i,entity in enumerate(entities_list):
  entity=re.sub(regex_delete_number,'',entity)
  entity=re.sub(regex_del_title,'',entity)
  entity=re.sub(regex_del_quotion_mark_start,'',entity)
  entity=re.sub(regex_del_quotion_mark_end,'',entity)

  entity=re.sub(regex_del_quotion_period,'',entity)
  entity=re.sub(regex_del_quotion_dash,'',entity)


  entities_list[i]=entity.strip()
```

```python
import pandas as pd
import os
import openai
import requests
import re
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file


def get_completion_from_messages(messages,
                                 model="gpt-3.5-turbo",
                                 temperature=0,
                                 max_tokens=500):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature,
        max_tokens=max_tokens,
    )
    return response.choices[0].message["content"]
```
