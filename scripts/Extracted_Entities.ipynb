```python
# Importing libraries

import pandas as pd
import re
```

```python
# Reading the both the .txt files and storing them in two lists

with open('Entities_0_400.txt') as f:
    entities_set1 = f.readlines()
with open('Entities_400_1327.txt') as f:
    entities_set2 = f.readlines()
```

```python
# Creating a dataframe of entities using the two lists

df_entities = pd.DataFrame(entities_set1+entities_set2, columns =['raw_entity'])
df_entities
```

```python
# Defining a function to process the entities to remove special charaters and clean the dataframe

def process_entity(raw_entity):
    raw_entity = re.sub("[\[\]\'\-\\n]", '', raw_entity)
    return raw_entity

df_entities["entity"] = df_entities["raw_entity"].apply(process_entity)
df_entities.drop('raw_entity', inplace=True, axis=1)
df_entities
```

```python
# Reading the question and answers csv file as a dataframe

df_que_ans = pd.read_csv("https://raw.githubusercontent.com/atreish/yukatzbot/main/tests/questions_answers_cleaned%20.csv")
df_que_ans
```

```python
# Merging the two dataframes - questions_answers and entities

df = pd.concat([df_que_ans,df_entities],axis=1)
df
```

```python
# Saving the final dataframe of questions, answers and entities as a csv file

df.to_csv("questions_answers_entities.csv", index=False)
```
